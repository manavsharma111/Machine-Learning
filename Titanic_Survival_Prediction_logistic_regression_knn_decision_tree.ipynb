```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(["Cabin","Pclass","Ticket","Name"], axis=1, inplace=True, errors='ignore')
```

```python
df.info()
```

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
df.dropna(subset=["Embarked"], inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.head()
```

```python
df["Sex"]=le.fit_transform(df["Sex"])
df["Embarked"]=le.fit_transform(df["Embarked"])
```

```python
df.drop(["Embark"], axis=1, inplace=True)
```

```python
df.head()
```

```python
X=df.drop("Survived", axis=1)
Y=df["Survived"]
```

LOGISTIC REGRASSION

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
model=LogisticRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
Y_pred
```

```python
Y_test
```

```python
confusion_matrix(Y_test, Y_pred)
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy Score: {accuracy:.4f}")
```

```python
from sklearn.metrics import precision_score
precision = precision_score(Y_test, Y_pred)
print(f"Precision Score: {precision:.4f}")
```

```python
from sklearn.metrics import recall_score, f1_score

recall = recall_score(Y_test, Y_pred)
f1 = f1_score(Y_test, Y_pred)

print(f"Recall Score: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define the parameter grid for Logistic Regression
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs'],
    'penalty': ['l1', 'l2']
}

# Create a GridSearchCV object
# Note: 'l1' penalty is only supported by 'liblinear' solver
# 'lbfgs' solver only supports 'l2' penalty
# We'll handle this by adjusting the grid for each solver in GridSearchCV implicitly.
# GridSearchCV will automatically skip invalid combinations.

grid_search = GridSearchCV(LogisticRegression(max_iter=200), param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit GridSearchCV to the training data
grid_search.fit(X_train, Y_train)

# Print the best parameters and best score
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validation accuracy: {grid_search.best_score_:.4f}")

# Get the best model
best_model = grid_search.best_estimator_

# Make predictions with the best model
Y_pred_tuned = best_model.predict(X_test)

# Evaluate the tuned model on the test set
accuracy_tuned = accuracy_score(Y_test, Y_pred_tuned)
precision_tuned = precision_score(Y_test, Y_pred_tuned)
recall_tuned = recall_score(Y_test, Y_pred_tuned)
f1_tuned = f1_score(Y_test, Y_pred_tuned)

print(f"Tuned Model Accuracy: {accuracy_tuned:.4f}")
print(f"Tuned Model Precision: {precision_tuned:.4f}")
print(f"Tuned Model Recall: {recall_tuned:.4f}")
print(f"Tuned Model F1 Score: {f1_tuned:.4f}")
```
